# Least Squares Regression

## Introduction

Least squares regression is a common method used to fit a linear model to a set of data points. It aims to find the coefficients of a linear equation that minimizes the sum of the squared differences between the observed and predicted values.

## Linear Regression Equation

The linear regression equation takes the form:

$$
y = mx + b
$$

## Equation for Coefficients

To find the coefficients $ m $ and $ b $ that best fit the data, we use the method of ordinary least squares (OLS). The equations for calculating $ m $ and $ b $ using OLS, with summation notation, are:

The equations for $ m $ and $ b $ using summation notation are derived from the method of ordinary least squares, which minimizes the sum of squared differences between the observed and predicted values. The terms in the equations represent summations of products and differences between the data points, allowing us to calculate the slope $ m $ and y-intercept $ b $ that best fit the data.





```python
# Plot by manually calculating m and b

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)

m = 0.82143
b = 1.67857

# Plotting

# y = mx + b

plt.plot (x, y, "o")

plt.plot (x, m*x + b, "r")

plt.show()
```

### The formula to calculate the coefficients using OLS in computational implimentation is:

$$
\beta = (A^T A)^{-1} A^T y
$$

where:
- $\beta$ is the vector of coefficients,
- $A$ is the design matrix (matrix of predictor variables),
- $y$ is the column vector of dependent variable values,
- $(A^T A)^{-1}$ is the inverse of the matrix product of the transpose of $A$ and $A$, and
- $A^T$ is the transpose of $A$.




## Implementation Steps

1. **Construct the Design Matrix ($A$)**:
   - Stack the independent variable values ($x$) along with a column of ones to represent the constant term (y-intercept).

2. **Format the Dependent Variable ($y$)**:
   - Reshape the dependent variable vector into a column vector format.

3. **Calculate Regression Coefficients ($\beta$)**:
   - Use the formula $\beta = (A^T A)^{-1} A^T y$ to find the regression coefficients.

4. **Visualize the Results**:
   - Plot the original data points and the regression line ($y = mx + b$) to analyze the fit of the linear model.

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)

# Step 1

# np.stack --> creating vertical stack
# np.ones --> creating an array with value = 1
# .T --> Transpose

A = np.vstack([x, np.ones(len(x))]).T

# print (A)

# Step 2

# np.newaxis --> reating new axis

y = y[:,np.newaxis]

# print (y)

# Step 3

# np.dot --> do the product rule matrix
# np.linalg.inv --> to invers a matrix

beta = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

print (beta)

# index 0 --> m, index 1 --> b

m = beta[0]
b = beta[1]

# Step 4

plt.plot (x, y, "bo")

plt.plot (x, m*x + b, "r")

plt.show()
```

```python
# Pseudo-Inverse

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T

y = y[:,np.newaxis]

pinv = np.linalg.pinv(A)

beta = np.dot(pinv, y)

print (beta)

m = beta[0]
b = beta[1]

plt.plot (x, y, "bo", label = "Data Points")

plt.plot (x, m*x + b, "r" , label = "Fitted Regression Line")

plt.title ("Secret data :)")

plt.xlabel ("X values")
plt.ylabel ("Y values")

plt.legend()
plt.show()
```

# Linear Interpolation
- interp1d = Function from scipy library to interpolate between x and y

```python
from scipy.interpolate import interp1d as inter
```

```python
# Linear Interpolation

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

# creating f to make a linear interpolation function

f = inter(x, y)

print (f)

x_new = 2.5
y_new = f (x_new)

print (y_new)

plt.plot (x, y, "b")
plt.plot (x_new, y_new, "ro")

plt.show()
```

# Cubic Spline Interpolation
- CubicSpline = Function from scipy library to interpolate using cubic spline between x and y

```python
from scipy.interpolate import CubicSpline as cubic
```

```python
# Cubic Spline Interpolation

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

# creating to f to make a cubic spline function

f = cubic(x, y)

# linspace --> creating an array with the same gap

x_new = np.linspace (0, 10, 10000)
y_new = f (x_new)

plt.plot (x, y, "bo")
plt.plot (x_new, y_new, "r")

plt.show()
```

# Newton Polynomial Interpolation

### 2 Main Function
- Polyfit = Coefficient of the polynomial
- Polyval = Value of each coefficient

```python
# Newton Polynomial Interpolation

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array (x)
y = np.array (y)

plt.figure (figsize = (12, 40))

for i in range (1, 21):
    plt.subplot (10, 4, i)
    plt.plot (x, y, "bo")

    y_new = np.polyfit (x, y, i)

    plt.plot (x, np.polyval(y_new, x), "r")

    plt.title (f"Polynomial Order {i}")

plt.show()
```
